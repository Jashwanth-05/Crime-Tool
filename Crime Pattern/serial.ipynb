```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset from the CSV file
df = pd.read_csv('skd.csv')

# Convert Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Define colors for gender categories
gender_colors = {'Female': '#1f77b4', 'Male': '#ff7f0e'}

# Plot temporal patterns
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Date', hue='Date')
plt.title('Temporal Patterns of Serial Killings')
plt.xlabel('Date')
plt.ylabel('Count of Killings')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot spatial patterns
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Location', hue='Location', palette='viridis')
plt.title('Spatial Patterns of Serial Killings')
plt.xlabel('Location')
plt.ylabel('Count of Killings')
plt.legend(title='Location', labels=df['Location'].unique())  # Set legend labels
plt.tight_layout()
plt.show()

# Plot victim demographics
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Victim Age', bins=10, kde=True, hue='Victim Gender', palette=gender_colors)
plt.title('Distribution of Victim Age by Gender')
plt.xlabel('Victim Age')
plt.ylabel('Count')
plt.legend(title='Victim Gender', labels=['Female', 'Male'])  # Set legend labels
plt.tight_layout()
plt.show()

# Plot methods of killing
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Method of Killing', hue='Method of Killing', palette='viridis')
plt.title('Distribution of Methods of Killing')
plt.xlabel('Method of Killing')
plt.ylabel('Count of Killings')
plt.xticks(rotation=45)
plt.legend(title='Method of Killing', labels=df['Method of Killing'].unique())  # Set legend labels
plt.tight_layout()
plt.show()
```
